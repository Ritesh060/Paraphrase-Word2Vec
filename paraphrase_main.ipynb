```python
import gensim
from gensim.models import Word2Vec

new_model = Word2Vec.load('model_final.bin')
print(new_model) 
```

```python
words = list(new_model.wv.index_to_key)
words 
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.corpus import stopwords    
import re

nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')

def filter_words(sentence):
    stop_words = set(stopwords.words('english'))

    pos_tags = ['DT', 'NN', 'NNS', 'NNP', 'NNPS']


    sentence = re.sub("[^A-Za-z]+", " ", sentence)
    sentence = re.sub(r'\$\w*', '', sentence)
    words = word_tokenize(sentence.lower())

    tagged_words = pos_tag(words)

    filtered_words = [word for word, tag in tagged_words if tag not in pos_tags and word.lower() not in stop_words]

    return filtered_words
```

```python
def find_similar_words(filtered_words):
    similar_words = []

    for word in filtered_words:
        if word in words:
            sim_words = new_model.wv.most_similar(word, topn=1)
            similar_words.append((word, sim_words[0][0]))
            
    similar_words = dict(similar_words)
    return similar_words
```

```python
def final_paraphrase(sentence, filtered_words, similar_words):
    new_sentence = sentence.split()
    final = ""
    l = len(new_sentence)
    for i in range(l):
        if new_sentence[i] in filtered_words:
            new_sentence[i] = similar_words.get(new_sentence[i])
        final = final + " " + new_sentence[i]
    return final
```

```python
def grammar_check(final):
    from gingerit.gingerit import GingerIt
    parser = GingerIt()
    f = parser.parse(final)
    return f['result']
```

```python
sentence = str(input("Enter a string to paraphrase: "))
f = filter_words(sentence)
f1 = find_similar_words(f)
final = final_paraphrase(sentence, f, f1)
grammar_check(final)
```
